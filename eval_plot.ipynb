```python
import os

import numpy as np
import pandas as pd
import plotly.express as px
```

```python
baseline_path = "/root/sae-transfer-learning/eval/rs-post_all.csv"
```

```python
baseline = pd.read_csv(baseline_path)
baseline.rename(columns={'Unnamed: 0':'L'}, inplace=True)
baseline
```

```python
baseline[["ACT-Layer", "SAE-Layer"]] = baseline["L"].str.split("-", n=1, expand=True)
baseline["ACT-Layer"] = baseline["ACT-Layer"].astype(int)
baseline["SAE-Layer"] = baseline["SAE-Layer"].astype(int)
baseline
```

```python
# Use plotply to plot a heatmap out of metrics[['ce_loss_score']].to_numpy().reshape(12, 12)
fig = px.imshow(
    baseline[["ce_loss_score"]].to_numpy().reshape(12, 12),
    color_continuous_scale="RdBu",
    labels=dict(x="SAE Layer", y="ACT Layer", color="CE Loss Score"),
    zmin=-1,
    zmax=1,
)
fig.update_layout(
    xaxis=dict(
        tickmode="array",
        tickvals=list(range(12)),
        ticktext=list(range(12)),
    ),
    yaxis=dict(
        tickmode="array",
        tickvals=list(range(12)),
        ticktext=list(range(12)),
    ),
)
fig.show()
```

```python
transfer_metric_path = "/root/sae-transfer-learning/eval/rs-post_transfer_forward_final_500002816_partial.csv"
transfer_metrics = pd.read_csv(transfer_metric_path)
```

```python
transfer_metrics.rename(columns={"Unnamed: 0": "L"}, inplace=True)
transfer_metrics
```

```python
transfer_metrics[["ACT-Layer", "SAE-Layer"]] = transfer_metrics["L"].str.split("-", n=1, expand=True)
transfer_metrics["ACT-Layer"] = transfer_metrics["ACT-Layer"].astype(int)
transfer_metrics["SAE-Layer"] = transfer_metrics["SAE-Layer"].astype(int)
transfer_metrics
```

```python
len(metrics[metrics["ACT-Layer"] == (metrics["SAE-Layer"] + 1)]["ce_loss_score"].to_numpy().flatten().tolist()[:-1])
```

```python
plot_df = pd.DataFrame(
    np.asarray(
        [
            transfer_metrics[["ce_loss_score"]].to_numpy()[0::2].flatten().tolist(),
            transfer_metrics[["ce_loss_score"]].to_numpy()[1::2].flatten().tolist(),
            metrics[metrics["ACT-Layer"] == metrics["SAE-Layer"]]["ce_loss_score"].to_numpy().flatten().tolist()[:-1],
            metrics[metrics["ACT-Layer"] == (metrics["SAE-Layer"] + 1)]["ce_loss_score"].to_numpy().flatten().tolist(),
        ]
    ).T, columns=["Transfer: SAE[i] -> ACT[i]", "Transfer: SAE[i] -> ACT[i+1]", "Baseline: SAE[i] -> ACT[i]", "Baseline: SAE[i] -> ACT[i+1]"]
)
```

```python
plot_df
```

```python
# Use plotply to plot a line chart out of transfer_metrics[['ce_loss_score']].to_numpy()[:, 0::2]
fig = px.line(
    plot_df.iloc[:-1],
    labels=dict(index="Layer", value="CE Loss Score", color="Transfer Type"), title="Transfer CE Loss Score: i-th SAE finetuned on (i+1)-th ACT",
)
fig.update_layout(
    xaxis=dict(
        tickmode="array",
        tickvals=list(range(11)),
        ticktext=list(range(11)),
    ),
)
fig.show()
```
